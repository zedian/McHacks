```python
import pandas as pd
import matplotlib as mpl
```

```python
dataF = pd.read_csv('training_SyncTranscript.csv')
dataL = pd.read_csv('training_SyncDiagnosis.csv')
```

```python
print(dataF.shape)
print(dataL.shape)
```

```python
print(dataF.head())
print(dataL.head())
```

```python
dataF = dataF.sort_values(by = 'PatientGuid')
print(dataF.shape)
```

```python
dataL = dataL.sort_values(by = 'PatientGuid')
print(dataL.shape)
```

```python
dataF.head()
```

```python
dataL.head()
```

```python
result = pd.concat([dataF, dataL], axis=1, sort=False)
results = pd.merge(dataF, dataL, on = 'PatientGuid')
```

```python
to_drop = ['TranscriptGuid', 'PatientGuid','VisitYear','RespiratoryRate','HeartRate','PhysicianSpecialty','UserGuid_x', 'DiagnosisGuid','ICD9Code','StartYear','StopYear','Acute','UserGuid_y']
results.drop(to_drop, inplace = True, axis = 1)
```
